**REMOVE** **NON** **ENGLISH** **WORDS**

```python
import langid

def remove_non_english_words(sentence):
    # Split the sentence into words
    words = sentence.split()

    # Define the languages to keep (English)
    languages_to_keep = {'en'}

    # Detect the language of each word and filter out non-English words
    english_words = [word for word in words if langid.classify(word)[0] in languages_to_keep]

    # Join the remaining English words back into a sentence
    english_sentence = ' '.join(english_words)

    return english_sentence
df['original_text'] = df['original_text'].apply(lambda text: remove_non_english_words(text))
df['original_text'].head()

```

```python
# Install necessary libraries
!pip install pandas numpy scikit-learn emoji gensim

# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, classification_report
import emoji
from gensim.models import Word2Vec
from scipy.sparse import hstack, csr_matrix  # Import csr_matrix

# Load your dataset
# url = "your_dataset.csv"
# df = pd.read_csv(url)

df3 = df.copy()

# Preprocess text data (you may need more preprocessing steps depending on your data)
df3['original_text'] = df3['original_text'].apply(lambda x: emoji.demojize(x))

# Train Word2Vec model on the text data
tokenized_text = [text.split() for text in df3['original_text']]
word2vec_model = Word2Vec(sentences=tokenized_text, vector_size=100, window=5, min_count=1, workers=4)



# Define the extract_emojis function using regular expressions
def extract_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)
    return emoji_pattern.findall(text)

# Train emoji2vec model on the text data
emoji_tokenized_text = [[word for word in text.split() if word in extract_emojis(text) or re.match(r'^\w+$', word)] for text in df3['original_text']]
emoji2vec_model = Word2Vec(sentences=emoji_tokenized_text, vector_size=100, window=5, min_count=1, workers=4)

# Modify the get_word_vector function
def get_word_vector(word):
    try:
        # Check if the word is an emoji
        if word in emoji2vec_model.wv.index_to_key:
            return emoji2vec_model.wv[word]
        else:
            return word2vec_model.wv[word]
    except KeyError:
        return np.zeros(100)  # Return zero vector for unknown words

# Apply the function to create a new column with word vectors
df3['word_vector'] = df3['original_text'].apply(lambda x: np.array([get_word_vector(word) for word in x.split()]))

# Filter out rows where there are no word vectors
df3 = df3[df3['word_vector'].apply(lambda x: len(x) > 0)]

# Combine text and word vectors
X_text = df3['original_text']
X_word = np.vstack(df3['word_vector'])

# Check if X_word is not empty
if X_word.size > 0:
    min_rows = min(X_text.shape[0], X_word.shape[0])
    X_combined = np.hstack([X_text.values[:min_rows].reshape(-1, 1), X_word[:min_rows]])
else:
    X_combined = np.array(X_text.values.reshape(-1, 1))

# Split the data into training and testing sets
y = df3['sentiment']
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)


# Use CountVectorizer to convert text data into numerical features (sparse matrix)
vectorizer = CountVectorizer(max_features=100000)
X_train_text_sparse = vectorizer.fit_transform(X_train[:, 0])
X_test_text_sparse = vectorizer.transform(X_test[:, 0])

# Convert word vectors to float and use sparse matrix representation
X_train_word_sparse = csr_matrix(X_train[:, 1:].astype(float))
X_test_word_sparse = csr_matrix(X_test[:, 1:].astype(float))

# Use sparse matrix representation
X_train_combined_sparse = hstack([X_train_text_sparse, X_train_word_sparse])
X_test_combined_sparse = hstack([X_test_text_sparse, X_test_word_sparse])


# Evaluate the models
def evaluate_model(classifier, X_test_combined, y_test):
    y_pred = classifier.predict(X_test_combined)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    return accuracy, report
```
